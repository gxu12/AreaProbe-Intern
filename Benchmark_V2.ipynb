```python
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from termcolor import colored
pd.set_option('display.max_columns', None)
```

```python
cleaned_hospital = pd.read_csv('cleaned_hospital_cleaned.csv')
```

```python
cleaned_hospital.drop(cleaned_hospital.columns[0:3],axis=1,inplace=True)
```

```python
cleaned_hospital.sample(5)
```

```python
hospital_reviews = pd.read_csv('~/downloads/all_hospital_reviews.csv',index_col=[0])
hospital_reviews.head()
```

```python
hospital_reviews['name_only'] = hospital_reviews['name_only'].mask(hospital_reviews['name_only']==0).fillna(hospital_reviews['loc_searched'].str.split(',').str[0]).reset_index(drop=True)
hospital_reviews = hospital_reviews[['name_only','avg_review']]
hospital_reviews = hospital_reviews.drop_duplicates().reset_index(drop=True)

avg_review = hospital_reviews.groupby('name_only').mean().reset_index()
```

```python
unique_name = avg_review['name_only'].unique().tolist()
cleaned_hospital['name_only'] = cleaned_hospital['infos.HospitalName'].apply(lambda x:''.join([part for part in unique_name if x in part]))
cleaned_hospital = cleaned_hospital.merge(avg_review, on='name_only')
```

```python
list_ = ['infos.TotalPatientDays','infos.TotalDischarges','infos.MedicareCertifiedBeds']
for value in list_:
    cleaned_hospital[value] = cleaned_hospital[value].astype(str).apply(lambda x:x.replace(',',''))
    
cleaned_hospital['Average_Length_Of_Stay'] = round(cleaned_hospital['infos.TotalPatientDays'].replace(',','').astype(float) / cleaned_hospital['infos.TotalDischarges'].astype(float),3)
cleaned_hospital['Occupancy Rates (%)'] = round((cleaned_hospital['infos.TotalPatientDays'].replace(',','').astype(float) / (cleaned_hospital['infos.MedicareCertifiedBeds'].astype(float) * 365)) * 100,3)
```

```python
def comparision_between_hospital(hospitalA,hospitalB):
    available_service_types = ['Beds','Infos','Clinical Services',
                               'Routine Services','Total Acute',
                               'Swing Beds','Total DPU',
                               'Total','Intensive Care','Psychiatric DPU',
                               'Nursery','Surgical Intensive Care',
                               'Neonatal Intensive Care','Labor and Delevery',
                               'Rehabilitation DPU','Pediatric Intensive Care',
                               'Skilled Nursing','Revenue']
    available_A = []
    available_B = []
    df_A = cleaned_hospital[cleaned_hospital['infos.HospitalName'].str.contains(hospitalA)]
    df_B = cleaned_hospital[cleaned_hospital['infos.HospitalName'].str.contains(hospitalB)]
    for value in available_service_types:
        selected_columns =  [col for col in cleaned_hospital.columns if value in col]
        selected_columns = list(selected_columns)
        if df_A[[*selected_columns]].isnull().all(axis=1).values == False:
            available_A.append(value)
        if df_B[[*selected_columns]].isnull().all(axis=1).values == False:
            available_B.append(value)
    
    basic_info_df_A = df_A[['Average_Length_Of_Stay','Occupancy Rates (%)','avg_review','infos.City','infos.State','infos.Zipcode']]
    basic_info_df_B = df_B[['Average_Length_Of_Stay','Occupancy Rates (%)','avg_review','infos.City','infos.State','infos.Zipcode']]
    basic_info_df = pd.concat([basic_info_df_A,basic_info_df_B],axis=0)
    basic_info_df.index = [hospitalA,hospitalB]
    
    print(colored("HOSPITAL COMPARISON",'red'))
    print("======================")
    print(colored('Available Service Types for ' + hospitalA + 'is:\n','yellow'))
    print(available_A)
    print(colored('\nAvailable Service Types for '+ hospitalB + 'is:\n','yellow'))
    print(available_B)
    print('\n')
    print("======================")
    type_of_service = str(input(' Please select one type of services \n\n'+colored('If not certain, please type "Not Sure"','red') +'\n\n Then all available types of services will show: \n'))
    if type_of_service.lower() == 'not sure':
        print('-----------------------')
        print(colored('AVAILABLE TYPES OF SERVICES ARE: \n','red'))
        print(colored('Beds','green')+': Information about number of beds')
        print(colored("Infos",'green')+": Hospital's information")
        print(colored('Clinical Services','green')+': Information of clinical services')
        print(colored('Routine Services','green')+': Information of routine services')
        print(colored('Total Acute','green')+': Information of acute in total')
        print(colored('Swing Beds','green')+': Information of post-hospital SNF care')
        print(colored('Total DPU','green')+': Information of total DPU(distinct part unit)')
        print(colored('Total','green')+': Information of total')
        print(colored('Intensive Care','green')+': Information of intensive care')
        print(colored('Psychiatric DPU','green')+": Psychiatric's DPU(distinct part unit)")
        print(colored('Nursery','green')+': Nursery information')
        print(colored('Surgical Intensive Care','green')+': Intensive care in Surgical')
        print(colored('Neonatal Intensive Care','green')+': Neonatal Intensive Care(Newborn baby)')
        print(colored('Labor and Delevery','green')+': process of childbirth')
        print(colored('Rehabilitation DPU','green')+': help the patient walk and speak clearly again')
        print(colored('Pediatric Intensive Care','green')+': specialty of medical science concerned with the physical, mental, and social health of children from birth to young adulthood')
        print(colored('Skilled Nursing','green')+': skilled nursing facility')
        print(colored('Revenue','green')+': Hospital Revenue')
        print('\n')
        type_of_service = str(input('Please Select one type of service: \n'))
    else:
        if not cleaned_hospital.filter(like=type_of_service,axis=1).shape[1]:
            print('\n')
            print('Please make sure you type rightly.')
            return comparision_between_hospital(hospitalA,hospitalB)
    selected_columns =  [col for col in cleaned_hospital.columns if type_of_service in col]
    selected_columns = list(selected_columns)    
    Hospital_A_df = cleaned_hospital[cleaned_hospital['infos.HospitalName'].str.contains(hospitalA)]
    Hospital_A_name = Hospital_A_df['infos.HospitalName'].values[0]
    Hospital_A_type_of_facility = (Hospital_A_name + '___' +
                                           Hospital_A_df['infos.TypeOfFacility'].reset_index(drop=True))
    Hospital_A_subset = Hospital_A_df[[*selected_columns]].T
    Hospital_A_subset.columns = Hospital_A_type_of_facility.rename('Type of Facility')

    Hospital_B_df = cleaned_hospital[cleaned_hospital['infos.HospitalName'].str.contains(hospitalB)]
    Hospital_B_name = Hospital_B_df['infos.HospitalName'].values[0]
    Hospital_B_type_of_facility = (Hospital_B_name + '___' +
                                           Hospital_B_df['infos.TypeOfFacility'].reset_index(drop=True))
    Hospital_B_subset = Hospital_B_df[[*selected_columns]].T
    Hospital_B_subset.columns = Hospital_B_type_of_facility.rename('Type of Facility')

    all_hospital = pd.concat([Hospital_A_subset,Hospital_B_subset],axis=1)
    
    print("======================")
    
    slice_row = input('Do you want to slice the row? \n' +
                      "'Y':Yes \n" +
                      "'N':No\n'")
    
    if slice_row == 'Y':
        number_of_rows = int(input('Please select the number of rows to show \n\n'))
    else:
        number_of_rows = all_hospital.shape[0]
    display(HTML('<h4>The Comparison between two hospitals</h4>'))
    
    try:
        fig,ax = plt.subplots()
        subset_cleaned_hospital = pd.concat([Hospital_A_df,Hospital_B_df],axis=0)
        beds_information = subset_cleaned_hospital.loc[:,subset_cleaned_hospital.columns.str.endswith('Beds')]
        beds_information = beds_information.dropna(axis=1,how='all').fillna(0)
        beds_information.index = all_hospital.columns
        ax = beds_information.plot(kind='bar',rot=0,ax=ax)
        max_width = 20
        ax.set_xticklabels(textwrap.fill(x.get_text(), max_width) for x in ax.get_xticklabels())
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    except TypeError:
        print(colored('There is no available bed information for both hospitals right now','yellow'))
    all_hospital = all_hospital.iloc[:number_of_rows]
    all_hospital = all_hospital.dropna(how='all')
    return display(basic_info_df),display(all_hospital.fillna('Not Available'))
    
    
```
